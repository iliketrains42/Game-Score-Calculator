```python
import bs4
from bs4 import BeautifulSoup
import requests
import html5lib
import pandas as pd
```

```python
url = "https://www.basketball-reference.com/leagues/NBA_2024_per_game.html"
text = requests.get(url).text
soup = BeautifulSoup(text, 'html5lib')
```

```python
d = pd.DataFrame()
d['Name'] = ""
d['Points'] = ""
d['Field Goals'] = ""
d['Field Goals Attempted'] = ""
d['Free Throws'] = ""
d['Free Throws Attempted'] = ""
d['Offensive Rebounds'] = ""
d['Defensive Rebounds'] = ""
d['Steals'] = ""
d['Assists'] = ""
d['Blocks'] = ""
d['Personal Fouls'] = ""
d['Turnovers'] = ""
d['Game Score'] = ""

```

```python
x = soup('tr', {'class', 'full_table'})
names = []
for i in x:
    z = i('a')
    for j in z:
        if '/players' in j['href']:
            names.append(j.string)
d['Name'] = names
```

```python
x = soup('tr', {'class', 'full_table'})
points = []
for i in x:
    z = i('td')
    for j in z:
        if j['data-stat'] == 'pts_per_g':
            points.append(float(j.string))
d['Points'] = points
```

Calculating Game Score. 
Game score is a metric that was developed by John Hollinger to quickly give a rough estimate of a player’s contribution to a game.


```python
d['Game Score'] = d['Points'] + 0.4*d['Field Goals'] - 0.7*d['Field Goals Attempted'] - 0.4*(d['Free Throws Attempted'] - d['Free Throws']) + 0.7*d['Offensive Rebounds'] + 0.3*d['Defensive Rebounds'] + d['Steals'] + 0.7*d['Assists'] + 0.7*d['Blocks'] - 0.4*d['Personal Fouls'] - d['Turnovers']
d.set_index(d['Name'], inplace = True)
```
